Change noisy speech and clean speech directory to desired wav files

```python
# Noisy speech and clean speech directory
noisy_speech = "../Dataset/dataset/clean/speech_1.wav"
clean_speech = "../Dataset/dataset/0dB/speech_1.wav"
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Generating the wideband spectrogram

# Load the noisy WAV file
file_path = noisy_speech
y, sr = librosa.load(file_path)

# Compute the Mel spectrogram
ms = librosa.feature.melspectrogram(
    y=y,
    sr=sr,
    S=None,
    n_fft=1024,
    hop_length=80,
    win_length=320,
    window='hann',
    center=True,
    pad_mode='constant',
    power=2.0
)

# Convert to dB scale
ms_db = librosa.power_to_db(ms, ref=np.max)

# Plot the spectrogram
plt.figure(figsize=(10, 6))
librosa.display.specshow(ms_db, sr=sr, n_fft=1024, hop_length=80, win_length=320, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title("Wideband Mel spectrogram")
plt.show()
```

```python
# Generating the narrowband spectrogram

# Load the noisy WAV file
file_path = noisy_speech
y, sr = librosa.load(file_path)

# Compute the Mel spectrogram
ms = librosa.feature.melspectrogram(
    y=y,
    sr=sr,
)

# Convert to dB scale
ms_db = librosa.power_to_db(ms, ref=np.max)

# Plot the spectrogram
plt.figure(figsize=(10, 6))
librosa.display.specshow(ms_db, sr=sr, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title("Narrowband Mel spectrogram")
plt.show()
```

```python
# Generating the clean speech as wideband spectrogram

# Load the clean WAV file
file_path = clean_speech
y, sr = librosa.load(file_path)

# Compute the Mel spectrogram
ms = librosa.feature.melspectrogram(
    y=y,
    sr=sr,
    S=None,
    n_fft=1024,
    hop_length=80,
    win_length=320,
    window='hann',
    center=True,
    pad_mode='constant',
    power=2.0
)

# Convert to dB scale
ms_db = librosa.power_to_db(ms, ref=np.max)

# Plot the spectrogram
plt.figure(figsize=(10, 6))
librosa.display.specshow(ms_db, sr=sr, hop_length=80, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title("Clean Speech at 9dB SNR")
plt.show()
```
